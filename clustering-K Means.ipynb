```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv(r'E:\ML\Sofia-\PPL.csv')
```

```python
data
```

```python
plt.scatter(data['CA100'],data['FT100'])
plt.xlabel('Continuous Assessment')
plt.ylabel('Formative Test')
plt.show()
```

```python
x=data.copy()
km=KMeans(4)
km.fit(x)
```

```python
clusters=x.copy()
clusters['cluster_pred']=km.fit_predict(x)
print(clusters)
```

```python
plt.scatter(clusters['CA100'],clusters['FT100'],c=clusters['cluster_pred'],cmap='rainbow')
plt.xlabel('Continuous Assessment')
plt.ylabel('Formative Test')
plt.show()
```
